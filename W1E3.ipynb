```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(E_J, E_L, theta, phi, phi_ext):
    return -2*E_J*np.cos(theta)*np.cos(phi + phi_ext/2) + E_L*phi**2
```

```python
E_J = 1
E_L = 0.01
phi_ext = 0
theta = np.linspace(-np.pi/2, 3/2*np.pi, 1000)
phi = np.linspace(-3*np.pi, 3*np.pi, 1000)
X, Y = np.meshgrid(theta, phi)
Z = f(E_J, E_L, X, Y, phi_ext)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10, 10))

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm,
                       linewidth=0, antialiased=False)

ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$\phi$')

plt.savefig('Gyenis_potential.png')

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(20, 14))
ax[0,0].plot(theta, f(E_J, E_L, theta, 0, 0))
ax[0,0].set_title(r'$\phi = 0, \phi_e = 0$')
ax[0,0].set_xlabel(r'$\theta$')

ax[0,1].plot(theta, f(E_J, E_L, theta, 0, np.pi*2))
ax[0,1].set_title(r'$\phi = 0, \phi_e = 2\pi$')
ax[0,1].set_xlabel(r'$\theta$')

ax[1,0].plot(phi, f(E_J, E_L, 0, phi, 0))
ax[1,0].set_title(r'$\theta = 0, \phi_e = 0$')
ax[1,0].set_xlabel(r'$\phi$')

ax[1,1].plot(phi, f(E_J, E_L, 0, phi, np.pi*2))
ax[1,1].set_title(r'$\theta = 0, \phi_e = 2\pi$')
ax[1,1].set_xlabel(r'$\phi$')


# Change the tick marks for ax[0,0]
ax[0,0].set_xticks([-np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2])
ax[0,0].set_xticklabels(['$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$'])

# Change the tick marks for ax[0,1]
ax[0,1].set_xticks([-np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2])
ax[0,1].set_xticklabels(['$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$'])

# Change the tick marks for ax[1,0]
ax[1,0].set_xticks([-3*np.pi, -2*np.pi, -np.pi, 0, np.pi, 2*np.pi, 3*np.pi])
ax[1,0].set_xticklabels(['$-3\pi$', '$-2\pi$', '$-\pi$', '0', '$\pi$', '$2\pi$', '$3\pi$'])

# Change the tick marks for ax[1,1]
ax[1,1].set_xticks([-3*np.pi, -2*np.pi, -np.pi, 0, np.pi, 2*np.pi, 3*np.pi])
ax[1,1].set_xticklabels(['$-3\pi$', '$-2\pi$', '$-\pi$', '0', '$\pi$', '$2\pi$', '$3\pi$'])

fig.savefig('Gyenis_potentials.png')
```
