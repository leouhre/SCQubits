```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1
E_C = 1
E_J = 4

def f(ratio, phi):
    # Consider only inductive potential, not kinetic charge energy
    # return 4*E_C*n**2 + E_J/24*phi**4 - E_J/2*phi**2 + 1/2*ratio*E_J*phi**2
    return E_J/24*phi**4 - E_J/2*phi**2 + 1/2*ratio*E_J*phi**2
```

```python
#plt.figure(figsize=(8, 6))

phi = np.linspace(-2*np.pi, 2*np.pi, 100)

for ratio in [0.5, 1, 2]:
    plt.plot(phi, f(ratio, phi), label=r'$\frac{\gamma}{N}=$'+str(ratio))

# Test to see contribution from negative quadratic term
#plt.plot(phi, -phi**2/2, label='Contribution from quadratic term')

plt.xlabel(r'$\phi$')
plt.ylabel(r'$E/h$ (GHz)')
plt.ylim(-2, 6)
plt.xlim(-1.5*np.pi, 1.5*np.pi)
plt.xticks([-1*np.pi, 0, 1*np.pi], [r'$-\pi$', r'$0$', r'$\pi$'])
plt.tick_params(axis='both', direction='in')
plt.legend()
plt.title('Potential energy of fluxionium qubit')
plt.savefig('fluxonium_potential_energy.jpg', dpi=300)
plt.show()
```

```python
# As the ratio increases, the quadratic term becomes more dominant, and the potential resembles that of a harmonic oscillator.
# In the other extreme, when the ratio is small, the potential is broadened beyond the quartic potential.
# The qubit transition frequencies of the latter will be smaller, and the anharmonicity increases as the potential differs from the quadratic potential.
```

```python
# For reference, we plot the function (24) without the cosine expansion to fourth order

def f2(ratio, phi):
    #return 4*E_C*n**2 -E_J*np.cos(phi) + 1/2*ratio*E_J*phi**2
    return - E_J*np.cos(phi) + 1/2*ratio*E_J*phi**2
```

```python
plt.figure(figsize=(10, 6))

phi = np.linspace(-np.pi, np.pi, 100)

for ratio in [0.5, 1, 2]:
    plt.plot(phi, f2(ratio, phi), label=r'$\frac{\gamma}{N}=$'+str(ratio))

plt.xlabel(r'$\phi$')
plt.ylabel(r'$H$')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

phi = np.linspace(-np.pi, np.pi, 100)

plt.plot(phi, f(1, phi), label='Approximation')
plt.plot(phi, f2(1, phi), label='Exact')

plt.xlabel(r'$\phi$')
plt.ylabel(r'$H$')
plt.legend()
plt.show()
```
